```python
import pandas as pd
```

```python
url = 'https://github.com/owid/covid-19-data/blob/master/public/data/owid-covid-data.csv?raw=true'
df = pd.read_csv(url, index_col=0)
df.head()
```

```python

# 1. Qué países se encuentran en el DataFrame
paises = df['location'].unique()
print(f"1. Países en el DataFrame:\n{paises}\n")
```

```python
# 2. Cuántos registros hay por continente
registros_por_continente = df['continent'].value_counts()
print(f"2. Registros por continente:\n{registros_por_continente}\n")
```

```python
# 3. Cuántos nuevos casos hubo como máximo en una fecha reportada por cada país
max_nuevos_casos = df.groupby('location')['new_cases'].max()
print(f"3. Máximos nuevos casos por país:\n{max_nuevos_casos}\n")
```

```python
# 4. Cuál es el país con el mayor GDP per cápita (PBI)
pais_max_gdp_per_capita = df.loc[df['gdp_per_capita'].idxmax(), 'location']
max_gdp_per_capita = df['gdp_per_capita'].max()
print(f"4. País con mayor GDP per cápita: {pais_max_gdp_per_capita} con {max_gdp_per_capita}\n")
```

```python
# 5. Qué país tiene la mayor cantidad de fumadoras femeninas
pais_max_fumadoras_femeninas = df.loc[df['female_smokers'].idxmax(), 'location']
max_fumadoras_femeninas = df['female_smokers'].max()
print(f"5. País con mayor cantidad de fumadoras femeninas: {pais_max_fumadoras_femeninas} con {max_fumadoras_femeninas}\n")
```

```python
# 6. Qué país tiene la mayor cantidad de muertes por millón
pais_max_muertes_por_millon = df.loc[df['total_deaths_per_million'].idxmax(), 'location']
max_muertes_por_millon = df['total_deaths_per_million'].max()
print(f"6. País con mayor cantidad de muertes por millón: {pais_max_muertes_por_millon} con {max_muertes_por_millon}\n")
```

```python
# 7. Qué país ha tomado las medidas más estrictas hasta ahora
pais_max_medidas_estrictas = df.loc[df['stringency_index'].idxmax(), 'location']
max_medidas_estrictas = df['stringency_index'].max()
print(f"7. País con medidas más estrictas: {pais_max_medidas_estrictas} con índice {max_medidas_estrictas}\n")
```

```python

# 8. Cuál fue el país en alcanzar los 10,000 casos activos por primera vez
df['date'] = pd.to_datetime(df['date'], errors='coerce')
df = df.dropna(subset=['date'])
df = df.sort_values(by='date')
df_paises_primeros_10000_casos_activos = df[df['total_cases'] >= 10000].groupby('location')['date'].min().sort_values()
pais_primeros_10000_casos_activos = df_paises_primeros_10000_casos_activos.idxmin()
fecha_primeros_10000_casos_activos = df_paises_primeros_10000_casos_activos.min()
```

```python
print(f"8. País que alcanzó los 10,000 casos activos primero: {pais_primeros_10000_casos_activos} (alcanzado el {fecha_primeros_10000_casos_activos})\n")
```
