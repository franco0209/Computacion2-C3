```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#1 del ejercicio 6
#Definimos la funcion del ejercicio 1 
def f(x):
    return np.sin(x)
#El rango 
x = np.linspace(-2, 2, 400)

# Plt para la funcion

plt.plot(x, f(x), label='f(x) = sin(x)')
plt.title('Ejercicio 6-1')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```

```python
# 3 del ejercicio 6 
def f(x):
    return 1/x

# Define the range
x = np.linspace(1, 10, 400)

# Plt para las funciones

plt.plot(x, f(x), label='f(x) = 1/x', color='green')
plt.title('Ejercicio 6 - 3')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```
