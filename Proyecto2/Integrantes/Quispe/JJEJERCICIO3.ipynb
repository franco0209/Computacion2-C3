```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt('covid.txt')
```

```python
n = len(data)
x = np.arange(n)
y = data

grado = 5
```

```python
 coeficientes = np.polyfit(x, y, grado)
```

```python
funcion_polinomica = np.poly1d(coeficientes)
x_fit = np.linspace(x.min(), x.max(), 100)
```

```python
y_fit = funcion_polinomica(x_fit)
```

```python
plt.plot(x, y, 'o', label='Datos originales')
plt.plot(x_fit, y_fit, label='Ajuste polin√≥mico')
plt.legend()
plt.show()
```
