```python
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import numpy as np
```

```python
# f(y,t)=dy/dt
def f(y, t):
    dydt = -2 * y * t
    return dydt

# C.I: y(0)=y_0
y0 = 1

# Dom t:0-5
j=0
t=[]
while j<=5:   
    t.append(j)
    j+=0.1

# Integrar la EDO
y = odeint(f, y0, t)

# Graficar y(n)
plt.plot(t, y)
plt.xlabel('t')
plt.ylabel('y(t)')
plt.title('Solución de la EDO')
plt.show()
```

```python
#Sol analítica: y(t)=e**-t**2
def yA(t):
    return np.e**-t**2
y2=[]
y2=list(map(yA,t))
# Graficar la solución
plt.plot(t, y2)
plt.xlabel('t')
plt.ylabel('y(t)')
plt.title('Solución analítica de la EDO')
plt.show()
```
