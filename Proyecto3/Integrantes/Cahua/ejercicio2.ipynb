```python
import numpy as np
import matplotlib.pyplot as plt
x=np.arange(-1,3,0.05)
y=np.zeros(len(x))
#print(x)
#print(len(x))
#print(y)
def f(x):
    return np.e**(-x**2)
for i in range(0,len(x)):
    y[i]=f(x[i])

y2=np.zeros(len(x))
for i in range(0,len(x)):
    y2[i]=1-f(x[i])
    
plt.plot(x,y,marker='o', linestyle='',color="red")
plt.plot(x,y2, linestyle='--',color="green")
plt.show()

```
