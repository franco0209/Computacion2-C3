```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-1, 3, 0.05)

y = np.zeros(len(x))
y2 = np.zeros(len(x))

# Definir la función f(x)
def f(x):
    return np.e**(-x**2)

# Calcular los valores de y y y2
for i in range(len(x)):
    y[i] = f(x[i])
    y2[i] = 1 - f(x[i])

# Crear el gráfico de dispersión para y
plt.scatter(x, y, color='blue', label='y = f(x)', alpha=0.6, edgecolors='w', linewidths=0.5)

# Crear el gráfico de dispersión para y2
plt.scatter(x, y2, color='red', label='y2 = 1 - f(x)', alpha=0.6, edgecolors='w', linewidths=0.5)

# Añadir etiquetas y título
plt.xlabel('x')
plt.ylabel('y / y2')
plt.title('Gráficos de Dispersión de f(x) y 1 - f(x)')
plt.legend()

# Mostrar el gráfico
plt.show()
```
