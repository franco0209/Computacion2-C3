```python
# Importamos las libreras necesarias para realizar clculos numricos y para la visualizacin
import numpy as np
import matplotlib.pyplot as plt 

# Definimos una funcin para crear y mostrar la grfica
def plot_frequencies(f1=10, f2=100, t=np.linspace(0, 0.5, 1000)):
    """
    Esta funcin genera una grfica de la seal compuesta por la suma de dos funciones seno con diferentes frecuencias.
    
    Parmetros:
    - f1: Frecuencia de la primera seal senoidal (default es 10 Hz)
    - f2: Frecuencia de la segunda seal senoidal (default es 100 Hz)
    - t: Arreglo de tiempo donde se evaluar la seal (default es un arreglo de 0 a 0.5 segundos con 1000 puntos)
    """
    
    # Calculamos la seal compuesta sumando dos senoidales con diferentes frecuencias
    signal = np.sin(2 * np.pi * f1 * t) + np.sin(2 * np.pi * f2 * t)
    plt.figure(figsize=(10, 5))
    
    # Graficamos la seal en funcin del tiempo, con la etiqueta 'Seal' y color rojo
    plt.plot(t, signal, label='Seal', color='red')
    plt.title('Dos frecuencias') 
    plt.xlabel('Tiempo (t)')
    plt.legend()
    plt.grid(True)
    plt.show()

# Llamamos a la funcin para generar y mostrar la grfica con los parmetros por defecto
plot_frequencies()
```
