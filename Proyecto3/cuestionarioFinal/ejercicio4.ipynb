```python
import numpy as np
import matplotlib.pyplot as plt
t=np.arange(-1,7,0.05)
g=np.zeros(len(t))
h=np.zeros(len(t))

for i in range(0,len(t)):
    g[i]=np.sin(t[i])
    
for i in range(0,len(t)):
    h[i]=np.cos(t[i])
    
fig, axs = plt.subplots(2)
fig.subplots_adjust(hspace=0.5, wspace=0.5)
axs[0].set_title("Función seno")
axs[1].set_title("Función coseno")
axs[0].plot(t,g)
axs[1].plot(t,h)
```
