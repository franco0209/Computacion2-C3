```python

#EJERCICIO 1



import numpy as np
import matplotlib.pyplot as plt
x=np.arange(-1,3,0.05)
y=np.zeros(len(x))
#print(x)
#print(len(x))
#print(y)
def f(x):
    return np.e**(-x**2)
for i in range(0,len(x)):
    y[i]=f(x[i])
 
plt.plot(x,y)
plt.show()
```
