```python
# Importamos las librerías necesarias para realizar cálculos numéricos y para la visualización
import numpy as np
import matplotlib.pyplot as plt 

# Definimos una función para crear y mostrar la gráfica
def plot_frequencies(f1=10, f2=100, t=np.linspace(0, 1000, 1000)):
    """
    Esta función genera una gráfica de la señal compuesta por la suma de dos funciones seno con diferentes frecuencias.
    
    Parámetros:
    - f1: Frecuencia de la primera señal senoidal (default es 10 Hz)
    - f2: Frecuencia de la segunda señal senoidal (default es 100 Hz)
    - t: Arreglo de tiempo donde se evaluará la señal (default es un arreglo de 0 a 0.5 segundos con 1000 puntos)
    """
    
    # Calculamos la señal compuesta sumando dos senoidales con diferentes frecuencias
    signal = np.sin(2 * np.pi * f1 * t) + np.sin(2 * np.pi * f2 * t)
    plt.figure(figsize=(10, 5))
    
    # Graficamos la señal en función del tiempo, con la etiqueta 'Señal' y color rojo
    plt.plot(t, signal, label='Señal', color='red')
    plt.title('Dos frecuencias') 
    plt.xlabel('Tiempo (t)')
    plt.legend()
    plt.grid(True)
    plt.show()

# Llamamos a la función para generar y mostrar la gráfica con los parámetros por defecto
plot_frequencies()
```
